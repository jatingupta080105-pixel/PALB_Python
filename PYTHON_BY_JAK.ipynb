write a program
str - INDIA
to reverse string without using the concept of slicing


```python
s = "INDIA"
reversed_s = ""
for char in s:
    reversed_s = char + reversed_s
print(f"Original string: {s}")
print(f"Reversed string: {reversed_s}")
```
